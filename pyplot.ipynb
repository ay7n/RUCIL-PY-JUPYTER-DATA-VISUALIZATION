## matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

**A simple example**

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = 2 + 3.0*x

plt.plot(x, y)
plt.show()
```

**How about feeding Python list?**

```python

#x = [0, 1, 2, 3, 4, 5]
y = 2 + 3.0*x

plt.plot(x, y)
plt.show()
```

**Saving figure to an image file**

```python
x = np.arange(0., 5., 0.1)
y = 2 + 3.0*x

plt.plot(x, y)
plt.savefig("st_line.png")
plt.draw()
```

**How to create subplots?**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(0, 10, 100)
y = x**2.0
plt.subplot(1,2,1)
plt.scatter(x,y)
plt.subplot(1,2,2)
plt.hist(x)
plt.show()
```

**Matrix of subplots**

```python
help(plt.hist)
```

```python
x = np.random.normal(0, 10, 100)
y = x**2.0
y1 = x**3.0
plt.subplot(2,2,1)
plt.scatter(x,y)
plt.subplot(2,2,2)
plt.hist(y, color='black')

plt.subplot(2,2,3)
plt.scatter(x,y1, color='green')
plt.subplot(2,2,4)
plt.hist(y1, color='red')
plt.show()
```

**Multiple Curves**

```python
x = np.arange(0., 5., 0.1)
y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x

plt.plot(x, y1, x, y2, x, y3)
plt.show()
```

**Multiple Curves**

```python
x = np.arange(0., 5., 0.1)
y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x


plt.plot(x, y1, marker="o") 
plt.plot(x, y2, marker="*") 
plt.plot(x, y3, marker="+") 
plt.title("Python experience with time")
plt.xlabel("time")
plt.ylabel("Productivity")
plt.show()
```

```python
x = np.arange(0., 5., 0.1)
y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x


plt.plot(x, y1, marker="o") 
plt.plot(x, y2, marker="*") 
plt.plot(x, y3, marker="+") 
plt.legend()
plt.show()
```

```python
x = np.arange(0., 5., 0.5)
y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x

plt.plot(x, y1, ls="dotted", label='line1') 
plt.plot(x, y2, linestyle="dashed",  label='line2') 
plt.plot(x, y3, ls=":",  label='line3') 
plt.legend(loc="upper left")
```
