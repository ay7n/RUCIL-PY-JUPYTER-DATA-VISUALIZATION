**A simple example**

### How to create subplots

```python
x = np.random.normal(0, 1, 1000)
y = x**2.0

plt.subplot(1,2,1)
plt.scatter(x,y)
plt.subplot(1,2,2)
plt.hist(x)
```

### Matrix of subplots

```python
help(plt.hist)
```

```python
x = np.random.normal(0, 1, 1000)
y = x**2.0
y1 = x**3.0

plt.subplot(2,2,1)
plt.scatter(x,y)
plt.subplot(2,2,2)
plt.hist(y, color='black')

plt.subplot(2,2,3)
plt.scatter(x,y1, color='green')
plt.subplot(2,2,4)
plt.hist(y1, color='red')
```

### Multiple Curves

```python
x = np.arange(0., 5., 0.1)

y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x

plt.plot(x, y1, x, y2, x, y3)
```

### Multiple Curves - add markers

```python
x = np.arange(0., 5., 0.1)
y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x


plt.plot(x, y1, marker="o") 
plt.plot(x, y2, marker="*") 
plt.plot(x, y3, marker="+") 
plt.legend()
```

```python
x = np.arange(0., 5., 0.5)
y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x

plt.plot(x, y1, ls="dotted", label='line1') 
plt.plot(x, y2, linestyle="dashed",  label='line2') 
plt.plot(x, y3, ls=":",  label='line3') 
plt.legend(loc="upper left")
```

```python
x = np.arange(0., 5., 0.1)
y1 = 2 + 1.0*x
y2 = 2 + 2.0*x
y3 = 2 + 3.0*x


plt.plot(x, y1, marker="o", label="electricity") 
plt.plot(x, y2, marker="*", label="water")
plt.plot(x, y3, marker="+", label="gas")
plt.legend(loc="upper left")
plt.title("Python experience with time")
plt.xlabel("time")
plt.ylabel("Productivity")
plt.show()
```

### Automatic vs Manual tick labels

```python
x = np.array([1, 2, 3, 4])
y = np.exp(x)
plt.plot(x, y)
plt.plot(x, y, marker="o" )
```

```python
x = np.array([1, 2, 3, 4])
y = np.exp(x)
plt.xticks([1, 2, 3, 4], ['one','two','three','four'])
plt.plot(x, y, marker="o" )
```
